### clean mails to remove missing values

```python
def get_rid_of_nulls(dataframe):
    for key in dataframe.keys():
        df[key] = df[key].apply(lambda x : x if not pandas.isnull(x) else 'is null value')
    return dataframe

get_rid_of_nulls(df)
df.head()
```

### Suppression des colonnes ayant des valeurs toutes identiques

```python
def remove_useless_col(dataframe):
    df_test = dataframe.copy()
    alleq = True
    for attribu in df_test:
        base = df_test[attribu][0]
        for donnee in df_test[attribu]:
            if(not donnee.__eq__('is null value') and not base.__eq__(donnee)):
                alleq = False
                break
        if(alleq): 
            del df_test[attribu]
    return df_test

df_test = remove_useless_col(df)
df_test.head()
```
